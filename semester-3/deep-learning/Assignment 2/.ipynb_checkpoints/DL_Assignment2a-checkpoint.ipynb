# 3. Data Preparation


```python
##---------Type the code below this line------------------##

# Load the dataset into a DataFrame
df_data_raw = pd.DataFrame(breast_cancer_wisconsin_diagnostic.data.features, columns=breast_cancer_wisconsin_diagnostic.variables['name'])

# Step 1: Eliminating Unnecessary Attributes
# Check if 'ID' column is present before dropping
columns_to_drop = ['ID', 'Diagnosis']
columns_to_drop = [col for col in columns_to_drop if col in df_data_raw.columns]
df_data = df_data_raw.drop(columns=columns_to_drop)

# Check if 'Diagnosis' column exists before dropping
if 'Diagnosis' in df_data.columns:
    # Drop the 'Diagnosis' column if all its entries are NaN
    if df_data['Diagnosis'].notna().all():
        df_data = df_data.drop(columns=['Diagnosis'])


df_target = pd.DataFrame(breast_cancer_wisconsin_diagnostic.data.targets, columns=['Diagnosis'])

# Concatenate the data and target DataFrames
df = pd.concat([df_data, df_target], axis=1)

# Define a dictionary mapping old column names to new column names
rename_dict = {
    'radius1': 'radius_mean',
    'texture1': 'texture_mean',
    'perimeter1': 'perimeter_mean',
    'area1': 'area_mean',
    'smoothness1': 'smoothness_mean',
    'compactness1': 'compactness_mean',
    'concavity1': 'concavity_mean',
    'concave_points1': 'concave_points_mean',
    'symmetry1': 'symmetry_mean',
    'fractal_dimension1': 'fractal_dimension_mean',
    'radius2': 'radius_se',
    'texture2': 'texture_se',
    'perimeter2': 'perimeter_se',
    'area2': 'area_se',
    'smoothness2': 'smoothness_se',
    'compactness2': 'compactness_se',
    'concavity2': 'concavity_se',
    'concave_points2': 'concave_points_se',
    'symmetry2': 'symmetry_se',
    'fractal_dimension2': 'fractal_dimension_se',
    'radius3': 'radius_worst',
    'texture3': 'texture_worst',
    'perimeter3': 'perimeter_worst',
    'area3': 'area_worst',
    'smoothness3': 'smoothness_worst',
    'compactness3': 'compactness_worst',
    'concavity3': 'concavity_worst',
    'concave_points3': 'concave_points_worst',
    'symmetry3': 'symmetry_worst',
    'fractal_dimension3': 'fractal_dimension_worst'
}


# Rename the columns using the dictionary
df.rename(columns=rename_dict, inplace=True)

# Separate continuous and categorical features
continuous_features = [col for col, dtype in df.dtypes.items() if dtype == 'float64']
categorical_features = [col for col, dtype in df.dtypes.items() if dtype == 'object']

# Checking and Fixing Missing Values
# Impute missing values for continuous features
imputer_continuous = SimpleImputer(strategy='mean')
df_continuous_imputed_values = imputer_continuous.fit_transform(df[continuous_features])
df_continuous_imputed = pd.DataFrame(df_continuous_imputed_values, columns=continuous_features, index=df.index)

# Impute missing values for categorical features
imputer_categorical = SimpleImputer(strategy='constant', fill_value='NA')
df_categorical_imputed_values = imputer_categorical.fit_transform(df[categorical_features])
df_categorical_imputed = pd.DataFrame(df_categorical_imputed_values, columns=categorical_features, index=df.index)

# Combine continuous and categorical DataFrames
df_imputed = pd.concat([df_continuous_imputed, df_categorical_imputed], axis=1)

## Split the data into training set and testing set
##---------Type the code below this line------------------##
X_train, X_test, y_train, y_test = train_test_split(df_continuous_imputed, y, test_size=0.3, random_state=42)

num_classes = 2

# Manually map class labels to integer values
class_mapping = {'B': 0, 'M': 1}

# Map class labels in y_train and y_test to integer values
y_train_encoded = y_train['Diagnosis'].map(class_mapping)
y_test_encoded = y_test['Diagnosis'].map(class_mapping)

# Convert integer-encoded labels to one-hot encoded format
y_train_one_hot = to_categorical(y_train_encoded, num_classes=num_classes)
y_test_one_hot = to_categorical(y_test_encoded, num_classes=num_classes)

## Identify the target variables.
##---------Type the code below this line------------------##
print(df_imputed.info())
```

## 4. Deep Neural Network Architecture


## 4.1 Design the architecture that you will be using

Journal refers two models RNN and GRU and done performace analysis for these two architectures. We will be modeling for these two artitectures


RNN model


```python
##---------Type the code below this line------------------##

# Define the RNN model
RNNmodel = Sequential([
    # Input layer
    SimpleRNN(units=128, return_sequences=True, input_shape=(30, 1)),
    Dropout(0.5),

    # RNN layer
    SimpleRNN(units=128, return_sequences=True),
    Dropout(0.5),

    # RNN layer
    SimpleRNN(units=64),

    # Dense layer
    Dense(units=4, activation='relu'),

    # Output layer
    Dense(units=2, activation='sigmoid')
])
```

GRU model

```python
##---------Type the code below this line------------------##

# Define the GRU model
GRUmodel = Sequential()

# Input Layer
GRUmodel.add(GRU(128, return_sequences=True, input_shape=(30, 1)))
GRUmodel.add(BatchNormalization())
GRUmodel.add(Dropout(0.5))

# GRU Layer 1
GRUmodel.add(GRU(128, return_sequences=True))
GRUmodel.add(BatchNormalization())
GRUmodel.add(Dropout(0.5))

# GRU Layer 2
GRUmodel.add(GRU(64))
GRUmodel.add(BatchNormalization())

# Dense Layers
GRUmodel.add(Dense(4))
GRUmodel.add(Dense(2))
GRUmodel.add(Activation('sigmoid'))
```

The choice of the number of parameters reflects the design decisions made by the model architect based on the complexity of the problem, the size of the dataset, and the desired model performance. The number of parameters impacts the model's capacity to learn complex patterns from the data. Too few parameters may lead to underfitting, while too many parameters may lead to overfitting. The specific numbers chosen for the parameters are based on empirical testing and optimization to balance model complexity and generalization ability.

```python
##---------Type the code below this line------------------##

# testing the RNN model for 150 epochs
testLoss_rnn_150, testAccuracy_rnn_150, testPrecision_rnn_150, testRecall_rnn_150, testF1Score_rnn_150 = RNNmodel.evaluate(X_test, y_test_one_hot)
```
