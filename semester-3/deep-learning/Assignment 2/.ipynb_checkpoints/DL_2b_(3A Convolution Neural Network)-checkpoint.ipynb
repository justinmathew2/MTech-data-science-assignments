# Convolutional Neural Network

```python
# Import Libraries
# - Tensorflow
# - Keras
# - numpy and random

import tensorflow as tf
from tensorflow.keras import models
from tensorflow.keras import layers

import random
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
random.seed(42)         # Initialize the random number generator.
np.random.seed(42)      # With the seed reset, the same set of numbers will appear every time.
tf.random.set_seed(42)  # sets the graph-level random seed
```

## Dataset - CIFAR10

```python
# Use the cifar10 dataset  of Keras.

(Xtrain, Ytrain), (Xtest, Ytest) = tf.keras.datasets.cifar10.load_data()

# Display size of dataset
print("Original shapes:")
print(Xtrain.shape, Xtest.shape)
print(Ytrain.shape, Ytest.shape)

# Normalize pixel values
Xtrain = Xtrain.astype('float32') / 255.0
Xtest = Xtest.astype('float32') / 255.0

# One-hot encode labels
Ytrain = tf.keras.utils.to_categorical(Ytrain, num_classes=10)
Ytest = tf.keras.utils.to_categorical(Ytest, num_classes=10)

# Reshape the dataset to match the model input shape
# CIFAR-10 images have shape (32, 32, 3)
# If you want to reshape to (28, 28, 1), you need to crop or resize the images, but it's not necessary for CIFAR-10
# So, we'll keep the original shape
Xtrain = Xtrain.reshape((50000, 32, 32, 3))
```

## Create a CNN Model

```python
# a) Conv Max Conv Max Conv Max Flatten FC FC
model_a = models.Sequential([
    layers.Conv2D(64, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (5, 5), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(256, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dense(256, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model_a.summary()

# b) Conv Conv Max Conv Conv Max Flatten Global Avg Pooling
model_b = models.Sequential([
    layers.Conv2D(64, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.Conv2D(128, (5, 5), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(256, (3, 3), activation='relu'),
    layers.Conv2D(512, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    #layers.Flatten(),
    layers.GlobalAveragePooling2D(),
    layers.Dense(10, activation='softmax')
])

model_b.summary()
```

```python
# Configure  the model for training, by using appropriate optimizers and regularizations
# Available optimizer: adam, rmsprop, adagrad, sgd
# loss:  objective that the model will try to minimize.
# Available loss: categorical_crossentropy, binary_crossentropy, mean_squared_error
# metrics: List of metrics to be evaluated by the model during training and testing.

model_a.compile(optimizer = "adam", loss = "categorical_crossentropy",
                 metrics = ["accuracy"])


model_b.compile(optimizer = "adam", loss = "categorical_crossentropy",
                 metrics = ["accuracy"])
```

```python
# train the model

model_a_history = model_a.fit(Xtrain, Ytrain, epochs = 10,
                       batch_size = 128, validation_split = 0.1)

```

```python
model_b_history = model_b.fit(Xtrain, Ytrain, epochs = 10,
                       batch_size = 128, validation_split = 0.1)
```

```python
print('Final training loss of model a\t', model_a_history.history['loss'][-1])
print('Final training accuracy of model a ', model_a_history.history['accuracy'][-1])
print('-----------------------------------------------------------------------')
print('Final training loss of model b\t', model_b_history.history['loss'][-1])
print('Final training accuracy of model b ', model_b_history.history['accuracy'][-1])
```

## Results and Outputs

```python
# testing the model_a

testLoss, testAccuracy = model_a.evaluate( Xtest, Ytest)
```

```python
print('Testing loss of model a \t', testLoss)
print('Testing accuracy model a', testAccuracy)
```

```python
# testing the model_b

testLoss, testAccuracy = model_b.evaluate( Xtest, Ytest)
```

```python
print('Testing loss of model b \t', testLoss)
print('Testing accuracy model b', testAccuracy)
```

```python
# shows the weights in layer 0 conv2d as gray map for model b
top_layer = model_b.layers[0]
plt.imshow(top_layer.get_weights()[0][:, :, :, 0].squeeze(), cmap='gray')
plt.show()
```

```python
# plotting training and validation loss for model a

model_a_loss = model_a_history.history['loss']
model_a_val_loss = model_a_history.history['val_loss']
epochs = range(1, len(model_a_loss) + 1)
plt.plot(epochs, model_a_loss, color='red', label='Training loss for model a')
plt.plot(epochs, model_a_val_loss, color='green', label='Validation loss for model a')
plt.title('Training and Validation loss for model a')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# plotting training and validation accuracy for model a

model_a_acc = model_a_history.history['accuracy']
model_a_val_acc = model_a_history.history['val_accuracy']
plt.plot(epochs, model_a_acc, color='red', label='Training acc for model a')
plt.plot(epochs, model_a_val_acc, color='green', label='Validation acc for model a')
plt.title('Training and Validation accuracy for model a')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# plotting training and validation loss for model b

model_b_loss = model_b_history.history['loss']
model_b_val_loss = model_b_history.history['val_loss']
epochs = range(1, len(model_b_loss) + 1)
plt.plot(epochs, model_b_loss, color='red', label='Training loss for model b')
plt.plot(epochs, model_b_val_loss, color='green', label='Validation loss for model b')
plt.title('Training and Validation loss for model b')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# plotting training and validation accuracy for model b

model_b_acc = model_b_history.history['accuracy']
model_b_val_acc = model_b_history.history['val_accuracy']
plt.plot(epochs, model_b_acc, color='red', label='Training acc for model b')
plt.plot(epochs, model_b_val_acc, color='green', label='Validation acc for model b')
plt.title('Training and Validation accuracy for model b')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

## Confusion Matrix generation

### Prediction for a specific testing data generte confusion matrix

```python
## For Model a

Y_prediction_a = model_a.predict(Xtest)

# Convert predictions classes to one hot vectors
Y_pred_classes_a = np.argmax(Y_prediction_a, axis = 1)

# Convert validation observations to one hot vectors
Y_true_a = np.argmax(Ytest,axis = 1)


## For model b

Y_prediction_b = model_b.predict(Xtest)

# Convert predictions classes to one hot vectors
Y_pred_classes_b = np.argmax(Y_prediction_b, axis = 1)

# Convert validation observations to one hot vectors
Y_true_b = np.argmax(Ytest,axis = 1)
```

```python
## Classification Report

from sklearn.metrics import classification_report

# For model a
print('FOR MODEL A')
print(classification_report(Y_true_a, Y_pred_classes_a))

# For model b
print('\n')
print('FOR MODEL B')
print(classification_report(Y_true_b, Y_pred_classes_b))
```

```python
### confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns

## compute the confusion matrix

# For model a
confusion_mtx_a = confusion_matrix(Y_true_a, Y_pred_classes_a)

plt.figure(figsize=(10,8))
sns.heatmap(confusion_mtx_a, annot=True, fmt="d");


# For model b
confusion_mtx_b = confusion_matrix(Y_true_b, Y_pred_classes_b)

plt.figure(figsize=(10,8))
sns.heatmap(confusion_mtx_b, annot=True, fmt="d");
```
