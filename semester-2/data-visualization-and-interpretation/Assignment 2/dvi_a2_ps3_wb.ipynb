### Question 1
<h4>What kind of a correlation exists between<br><br>
    a."engine-size" and "price"<br><br>
    b."highway-mpg" and "price"<br><br>
    c."peak-rpm" and "price"<br><br>
    Use appropriate visuals to demonstrate the same.Use subplots.
<h4>Clearly mention in markdown cells below the visual <br><br>1.You observataion about the relationship. Which among engine-size, highway-mpg and peak-rpm is a potential predictor variable of price.<br>
    2.Mention the pre-attentive attributes used.(atleast 2)<br>
                    3.Mention the gestalt principles used.(atleast 2)<br>
   

```python
# Subplot with 1 row and 3 columns
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Scatter plot and correlation for engine-size vs price
sns.scatterplot(x='engine-size', y='price', data=df, ax=axes[0])
corr_engine_size_price = df['engine-size'].corr(df['price'])
axes[0].set_title(f'Correlation: {corr_engine_size_price:.2f}', fontsize=12, color='green')

# Scatter plot and correlation for highway-mpg vs price
sns.scatterplot(x='highway-mpg', y='price', data=df, ax=axes[1])
corr_highway_mpg_price = df['highway-mpg'].corr(df['price'])
axes[1].set_title(f'Correlation: {corr_highway_mpg_price:.2f}', fontsize=12, color='red')

# Scatter plot and correlation for peak-rpm vs price
sns.scatterplot(x='peak-rpm', y='price', data=df, ax=axes[2])
corr_peak_rpm_price = df['peak-rpm'].corr(df['price'])
axes[2].set_title(f'Correlation: {corr_peak_rpm_price:.2f}', fontsize=12, color='red')

plt.tight_layout()
plt.show()
```

**Observation about the relationship**

*   There is a positive correlation between engine size and price. As engine size increases, the price tends to increase. This relationship appears to be moderately strong.
engine-size is a potential predictor variable of price.

*   The other two (highway-mpg and peak-rpm) seems to have negative correlation between them, which suggests that they may not be a potential predictor variable of price. In case of highway-mpg we can see that as highway mpg increases, the price tends to decrease.


**Pre-Attentive Attributes**


*   Position: The position of data points on the plot conveys the relationship between variables. The x-axis represents one variable, the y-axis represents another variable, and the position of each point indicates their values.

*   Color: The title of each subplot is color-coded to indicate the correlation coefficient. Positive correlations are displayed in one color, and negative correlations in another. This color coding helps draw attention to the strength and direction of the correlation.

**Gestalt Principles**


*   Proximity: The scatter plots are arranged in close proximity to each other in a single row, making it easy to compare the relationships between different pairs of variables.

*   Similarity: The scatter plots share a similar structure, with axes, data points, and titles. This similarity helps viewers quickly understand the purpose and content of each plot.

*   The type of drive wheels also seems to be related to car prices, rwd cars seems to have higher prices than fwd and 4wd. This suggests that drive-wheels could be a potential predictor variable of price.



**Pre-Attentive Attributes**


*   Position: The position of the boxes and whiskers on the y-axis of the boxplots is used to convey the distribution of prices within different categories of the categorical variables.

**Gestalt Principles**


*  Proximity: The boxplots for each categorical variable are arranged in close proximity to each other in a single row, making it easy to compare the relationships between different categories and prices.

*  Similarity: The boxplots share a similar structure with boxes and whiskers, which helps viewers quickly understand the distribution of prices within each category. This similarity aids in making comparisons between categories.

```python
# Set the number of rows and columns for subplots
num_rows = (len(obj_vars.columns) + 2) // 3
num_cols = 3

# Subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 20))
plt.subplots_adjust(wspace=0.4, hspace=0.4)

# Flatten the axes array
axes = axes.flatten()

# Bar plots for each categorical variable
for i, col in enumerate(obj_vars.columns):
    sns.countplot(data=df, y=col, palette='Set1', ax=axes[i])
    axes[i].set_xlabel('Count')
    axes[i].set_ylabel(col)
    axes[i].set_title(f'Distribution of {col}')

for j in range(len(obj_vars.columns), len(axes)):
    fig.delaxes(axes[j])

# Show the subplots
plt.show()
```

<h4> Use a single heat map to visualize the relationship between Body Style and Drive-wheel vs Price.</h4>
<h4>Perform the data preprocessing required.<br>

<p>The heatmap should plot  the target variable (price) proportional to colour with respect to the variables 'drive-wheel' and 'body-style' in the vertical and horizontal axis respectively. This allows us to visualize how the price is related to 'drive-wheel' and 'body-style'.</p>

The plot should have labels for x axis and y axis<br>
Write down your observations in a markdown cell below the visual</h4>

```python
# Pivoting the data to create a heatmap
heatmap_data = df.pivot_table(index='drive-wheels', columns='body-style', values='price', aggfunc='mean')

# Set the figure size
plt.figure(figsize=(12, 8))

# Creating a heatmap
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap="YlGnBu")

# Add labels to the axes
plt.xlabel('Body Style')
plt.ylabel('Drive Wheel')

# Set the title
plt.title('Heatmap: Price vs Body Style & Drive Wheel')

# Show the plot
plt.show()
```


**Observations**

*   From the heatmap, we can observe how the average price varies for different combinations of Body Style and Drive Wheel.

* It appears that convertible and hardtop body styles with rwd drive wheels tend to have higher average prices compared to other combinations.

*  Sedan body styles with fwd and rwd drive wheels have a wider range of prices.

**Pre-attentive attributes**

*   Color: Color is used to represent the average price, with lighter colors indicating higher prices and darker colors indicating lower prices. This allows viewers to quickly identify price patterns.

*   Orientation: The orientation of the heatmap, with Body Style on the vertical axis and Drive Wheel on the horizontal axis, allows viewers to understand the relationship between these categorical variables in the context of the price (continuous variable).

**Gestalt principles**
*   Similarity: Similar color shades are used for cells with similar price values, making it easier to group and compare them.

*   Continuity: The heatmap uses continuous gradients of color to represent the price values. This continuity in color transitions helps viewers understand the gradual changes in price across different combinations of Body Style and Drive Wheel. It enables viewers to perceive trends and variations in a smooth and continuous manner.

What is the distribution of car makes in the dataset, and how many cars does each make represent?

```python
# Number of cars for each make
make_counts = df['make'].value_counts()

# Set the figure size
plt.figure(figsize=(10, 6))

# Horizontal bar plot to visualize the distribution of car makes
sns.barplot(x=make_counts.values, y=make_counts.index, palette='Set1')

# Add labels to the axes
plt.xlabel('Number of Cars')
plt.ylabel('Car Make')

# Set the title
plt.title('Distribution of Car Makes')

# Show the plot
plt.show()
```

**Summary**

The bar plot displays the distribution of car makes in the dataset. It provides insights into how many cars each make represents. Some makes are more common, while others are less common. From the plot its clear that Toyota is the most common car make in the dataset, with 32 cars.

**Reason for selecting Horizontal Bar Plot**

**Pre Attentive Attributes**

*   Length: The length of each bar represents the number of cars for a particular make, making it easy to compare the counts across different makes.

*   Color: Different colors are used to distinguish between car makes, aiding in the identification of categories.
